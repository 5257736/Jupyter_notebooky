### Čo ty a basketbal?

Máš rád basketbal? Hral si už niekedy basketbal alebo sledoval si priebeh hry ako fanúšik? Ak áno, tak v tom príprade si na správnom mieste, kde sa o tejto hre naučíš, určite niečo nové. 

Pri nedovolenom bránení na strieľajúceho hráča počas hry, dostáva strelec možnosť strieľať 2 tresné hody. Táto situácia je štandardná herná situácia, ktorá je vymedzená pravidlami. Trestné hody sa strieľajú v stabilných podmienkach z konštantnej vzdialenosti 4,20 m. Počas trestného strieľania je hra prerušená. Taktiež na hráča nie je vyvíjaný tlak od súperov. Aj napriek tomu, že streľba trestných hodov sa vykonáva v štandardných podmienkach, tak je ovplyvnená mnohými faktormi (psychika, technika streľby, intenzita zaťaženia, prostredie atď.).

Zaujímavé to začína byť tu, keď sa budeme snažiť dokázať alebo vyvrátiť teóriu horúcej ruky. (Hráč premení prvý aj druhý trestný hod na bod.) Čiže, v tomto dokumente si zodpovedáme otázku, či druhý hod závisí od prvého hodu.

# <font color=blueviolet>Veriť alebo nie ?</font>


### Veria v horúcu ruku ľudia?

Vybrali sme 100 náhodných fanúšikov a opýtali sme sa ich jednoduchú otázku: ,,Má hráč pri trestnom strieľaní väčšiu šancu uspieť pri druhom hode, keď v prvom hode uspel, ako keď pri prvom strieľaní netrafil?". Až 68 respodentov odpovedalo **áno** (veria, že teória **horúca ruka** existuje) a 32 **nie**. Na modelovanie naznačenej situácie sme použili binomické rozdelenie $Bi(100,p)$. Zvolíme si, že $H_0:p\le\frac{1}{2}$. Chceli by sme, ale ukázať, že väčšina fanúšíkov verí v existenciu horúcej ruky, preto si zvolíme alternatívu $H_1:p>\frac{1}{2}$. Príkaz v R viď. nižšie:

```r
binom.test(68,100,0.5,alternative="greater")
```

Ako môžeme vidieť, tak $p$-hodnota je rovná $0.0002044$. To znamená, že nulovú hypotézu $H_0$ zamietame a pripúšťame platnosť alternatívnej hypotézy $H_1$.

### Zavedenie veličiny $Z$

Testovanie nezávislosti je založené na náhodnej veličine $Z$, o ktorej sa dá ukázať, že má asymptoticky (pre rastúce $n$) normálne rozdelenie. Najprv je však potrebné zaviesť náhodnú veličinu $X$ (reprezentuje to, že hráč trafil prvý aj druhý trestný hod) a jej strednú hodnotu - ozn. $\mathrm{E}(X)$, pričom platí: $\mathrm{E}(X)=\dfrac{n_{1.} n_{.1}}{n}$. Tiež budeme potrebovať varianciu náhodnej veličiny, ozn. $\mathrm{D}(X)$ resp. ($\mathrm{var}(X)$), pre varianciu (disperziu) náhodnej veličiny  platí: $\mathrm{D}(X)=\dfrac{n_{1.} n_{2.} n_{.1} n_{.2}}{(n-1)n^2}$. Teraz už máme všetky potrebné veličiny pre zavedenie náhodnej veličiny $Z$, ktorá je daná vzťahom: $Z=\dfrac{X-\mathrm{E}(X)}{\sqrt{\mathrm{D}(X)}}$. $Z$ má asymptoticky normálne rozdelenie $N(0,1)$. Vďaka náhodnej veličine $Z$, dokážeme testovať nezávislosť. Navyše podľa znamienka dostávame informáciu o smere prípradnej závislosti.

Máme Tab. 6: Výsledky hráčov z hľadiska závislostí. Teraz sa pozrieme na dáta hráčov, ktorí strieľali trestné hody. 

$$ \text{Tab. 6: Výsledky hráčov z hľadiska závislostí}$$

Žiadna z vypočítaných veličín $Z$ nie je štatisticky signifikatná, čo zamieta hypotézu o existencii horúcej ruky. Ale ako vidíme, tak 4 hodnoty sú kladné a 5 hodnôt je záporných. Z tejto úvahy môžeme opäť uvažovať o existencii horúcej ruky. Potvrdzuje sa, že fanušíkovia si pamätajú celkové štatistiky.

### Zhodnotenie celej problematiky

Porovnali sme rôzne situácie a skúmali sme, či existuje horúca ruka alebo nie. Zistili sme, že ak zoberieme štatistiky jednotlivých hráčov, tak existencia horúcej ruky sa zamieta. Ale keď vezmeme sumárne (spojené) dáta hráčov, tak tam existenciu horúcej ruky nemôžeme zamietnuť, preto je potrebné dávať pozor, aké dáta berieme do úvahy.
