### Kolky alebo bowling?

Kolky alebo bowling? Sú to rovnaké alebo rôzne hry? Veľa ľudí si myslí, že je to jedna a tá istá hra. Čo samozrejme **nie je** pravda.  
Bowling sa od koliek líši vo viacerých aspektoch. V kolkách sa používa 9 koliek, ktoré sú usporiadané do kosoštvorca, zatiaľ čo v bowlingu je rozmiestnených **10 koliek** do tvaru trojuholníka. Ďalšou odlišnosťou je, že kolky sa hrajú s guľami rovnakej hmotnosti bez dier. V bowlingu máme **rôzne hmotnosti gúľ**, ktoré majú **tri diery** na prsty a palec. Najzákladnejším rozdielom týchto hier je maximálne získaný počet bodov. V kolkách môžeme maximálne získať 120 bodov, zatiaľ, čo v bowlingu je to až **300 bodov**.

### Vieme pravidlá bowlingu?

Cieľom hráča je zhodiť vrhnutou bowlingovou guľou všetkých 10 koliek, ktoré stoja vo vzdialenosti 18,5 m rozostavané do rovnoramenného trojuholníka. Hmotnosť bowlingovej gule sa pohybuje od 3,6 do 7,2 kg. Bowling sa hrá na body a vyhráva ten, kto v hre dosiahne najvyššie bodové skóre.  
V každej hre sa hádže, v desiatich kolách po dva hody. V prípade, že hráč prvým hodom zhodí všetky kolky, druhý hod v danom kole už nehrá. **Strike** nastáva pri situácií, keď prvým hodom hráč zhodí všetkých 10 koliek. Ak prvým hodom hráč **nezhodí všetkých 10**, ale druhým hodom zhodí zvyšné, tak vtedy hovoríme, že nastal **spare**. Ak zhodí všetkých 10 v poslednom kole, hráč má bonusové hody (jeden v prípade spare a dva v prípade strike).  
Bowling sa hrá na dráhe, ktorá sa člení na tri časti:   

**1. kĺzajúcu;**  


**2. rotujúcu;**  


**3. oblúkovú.**

Poďľa pomenovaní vieme, ako sa správa bowlingová guľa na dráhe. V prvej časti, ktorá je najviac naolejovaná, guľa kĺže, v druhej sa vytvára jej rotácia v tretej, ktorá musí byť úplne suchá sa vytvára oblúk.


### Ako je to s bodmi?
Pri **strike** hráč získa **10 bodov + prémiu v hodnote jeho nasledujúcich dvoch hodov**. Ak by **3-krát** po sebe získal strike, tak za prvé kolo by získal 30 bodov. **Perfektná hra sa skladá z 12 strike s maximálnym počtom 300 bodov**.  

Príklad:  
$\text{1. kolo, 1. guľa: 10}$ 
  
$\text{2. kolo, 1. guľa: 10}$
  
$\text{3. kolo, 1. guľa: 10}$   
  
$\text{4. kolo, 1. guľa: 3}$   
  
$\text{4. kolo, 2. guľa: 6}$    

Za prvé kolo získa **30 bodov**, za 2. **23**, za 3. **19** a za 4. **9 bodov**, spolu teda **81 bodov**.  

Zhodením všetkých 10 koliek dvoma hodmi, čiže nastal **spare**, vtedy hráč získa **10 bodov + prémiu za prvý hod nasledujúceho kola**.   

Za dobrú hru sa považuje skóre nad **200** bodov.

### Závislé vs. nezávislé
Budeme predpokladať **strike** **(S)** za úspech a každý iný výsledok za neúspech **(N)**. Našou otázkou, záhadou je, že či výsledky hráča v priebehu hry sú na sebe nezávislé a či sa v priebehu hry nemení pravdepodobnosť úspechu.  
Úsek rovnakých výsledkov, ktoré sú ohraničené zľava aj sprava iným výsledkom sa nazýva **iterácia**. Tento pojem bude pre nás v tejto časti veľmi dôležitý.

# <font color=blueviolet> Použité testy </font>


### Vzorový príklad
Na tomto príklade si ukážeme testy, ktoré použijeme neskôr na reálnych dátach.  
Máme výsledok $$N,N,S,S,N,S,S,S,S,S,N$$

Použijeme:  
1. Test založený na **kontingenčnej tabuľke**:


$$\text{Tab. 1: Kontigenčná tabuľka}$$ 

Pomocou softvéru **R** otestujeme hypotézu nezávislosti v Tab. 1: takto:

```r
tbl<-matrix(c(5,2,2,1), byrow=T, nrow=2)
fisher.test(tbl, alternative="greater")
```

Vzhľadom na to, že **p-hodnota** je rovná $0.7083$, čo je vcelku vysoké číslo, tak $H_0$ nie je možné zamietnuť. ($H_0$ predpokladá, že výsledky sú na sebe nezávislé).

2. Test založený na **počte iteracií**:

```r
library(randomizeBE)
pruns.exact(r=5,n1=7,n2=4, tail="lower")
```

Vo funkcii `pruns.exact()` je $r$ rovno počet iteracií, $n1=a$ a $n2=b$, kde platí, že $a\ge b$. Máme tu prvky dvoch typov $S$ a $N$. Početnosť jedného typu oznacíme $a$, počet druhého typu označíme $b$. 
Tu je **p-hodnota** vypočítaná presne.

3. **Asymptotický** test: 

```r
library(tseries)
x<-factor(c("N","N","S","S","N","S","S","S","S","S","N"))
runs.test(x, alternative="less")
```

Pri malej dĺžke postupnosti je výraznejší rozdiel medzi **presnou p-hodnotou** a **asymptotickou p-hodnotou**. Hoci sa p-hodnoty aplikovaných testov nerovnajú, tak všetky testy vedú k **nezamietnutiu $H_0$**. Ak by p-hodnota vyšla menšia ako hladina významnosti testu $\alpha$ (štandardne $\alpha=0.05$), tak by sme mohli zamietnuť $H_0$ a prijať alternatívnu hypotézu $H_1$, ktorá hovorí o závislosti výsledkov hráča v priebehu hry.

# <font color=blueviolet> Aplikácia testov na nami získané reálne dáta </font>


### 1.PRÍKLAD
Začneme od najmenšieho bodového zisku.   
Máme výsledok $$N,N,N,N,N,S,S,N,S,N,N,N$$


$$\text{Tab. 2: Kontigenčná tabuľka}$$

```r
#1.spôsob
library(tseries)
x<-factor(c("N","N","N","N","N","S","S","N","S","N","N","N"))
runs.test(x,alternative="less")
```

```r
#2.spôsob
library(randomizeBE)
pruns.exact(r=5,n1=9,n2=3,tail = "lower")
```

```r
#3.spôsob
#prvý riadok - prvý stĺpec, prvý riadok - druhý stĺpec
tbl<-matrix(c(1,2,2,6),byrow=T, nrow=2)
fisher.test(tbl,alternative = "greater")
```

Ako môžeme vidieť $p$-hodnota je väčšia ako hladina $\alpha$, preto nemôžeme zamietnuť $H_0$.

### 2.PRÍKLAD   
Máme výsledok $$S,N,N,N,N,S,N,S,S,S,N,N$$

$$\text{Tab. 3: Kontigenčná tabuľka}$$

```r
#1. spoôsob
library(tseries)
z<-factor(c("S","N","N","N","N","S","N","S","S","S","N","N"))
runs.test(z,alternative="less")
```

```r
#2. spoôsob
library(randomizeBE)
pruns.exact(r=4,n1=7,n2=5,tail = "lower")
```

```r
#3. spoôsob
#prvý riadok - prvý stĺpec, prvý riadok - druhý stĺpec
tbl<-matrix(c(2,3,2,4),byrow=T, nrow=2)
fisher.test(tbl,alternative = "greater")
```

Opäť nemôžeme zamietnuť $H_0$, pretože $p$-hodnota je väčšia než $\alpha$.

### 3.PRÍKLAD

```r
#1. spôsob
library(tseries)
y<-factor(c("S","N","S","S","S","N","N","N","S","S","S","S"))
runs.test(y,alternative="less")
```

### Zhodnotenie celej problematiky


V tejto časti sme skúmali, či výsledky hráčov v priebehu hry sú od seba nezávislé. Išpirovali sme sa vzorovým príkladom. Následne sme na získaných dátach od hráča zrealizovali všetky tri testy. Pozorovali sme tiež, či dobre zahraté kolo (zahral strike) ovplyvní, že aj nasledujúce kolo bude dobre zahraté (zahrá strike opäť). 

Nezávislosť výsledkov berieme ako nulovú hypotézu $H_0$. Keďže sme vykonali viacero testov a ani jeden test nám nezamietol $H_0$ (čiže $p$-hodnota väčšia ako $\alpha$), takže sme nepotvrdili závislosť výsledkov. V žiadnom prípade nevyšla $p$-hodnota menšia ako $\alpha$. Preto nemôžeme potvrdiť, že jednotlivé kolá sú od seba závislé. Celá táto situácia je podobná situácií **horúcej ruky**. Horúca ruka v basketbale je v nasledujúcej kapitole rozanalyzovaná. V krátkosti je jedná o to, že či pri hodoch v trestnom strieľaní je závislosť. Konkrétnejšie vysvetlené, že či prvý premenený trestný hod ovplyvní druhý trestný hod. Podobná situácia len v bowlingu bola skúmaná v tejto kapitole. Nepotvrdili sme závislosť medzi jednotlivými kolami.
