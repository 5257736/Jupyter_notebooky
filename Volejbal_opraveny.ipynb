### Volejbal, áno alebo nie? Závislosť, áno alebo nie?

Určite si už počul o tomto olympijskom športe. Volejbal je populárny kolektívny šport, ktorý sa hrá na vonkajšom alebo vnútornom ihrisku. Hrá sa medzi dvoma družstvami, pričom každé družstvo má šesť hráčov na ihrisku. Cieľom hry je poslať loptu cez sieť na súperovu stranu tak, aby padla na zem do ihriska alebo aby ju súper nedokázal vrátiť. Často nastáva vo volejbale šnúra (resp. veľakrát za sebou skóruje jeden tím). Vtedy sa snaží tréner tímu, ktorý nemá šnúru, prerušiť hru.  
Čo myslíte, bude medzi týmito dvomi aspektami nejaká závislosť? Bude mať vplyv prerušenie hry trénerom alebo nie? Poďme si to spoločne rozanalyzovať!

### Na začiatok stručné pravidlá

Volejbalový zápas sa skladá zo setov. Všetky sety pozostávajú z výmen. Každá výmena sa začína podaním, pričom hráč z jedného tímu podáva loptu cez sieť na stranu druhého tímu. Ak sa lopta dotkne zeme ihriska alebo súper nezvládne spracovať podanie, bod získava podavajúci tím a hráč, ktorý podával, opäť podáva (servisuje). Ak lopta zostane v hre, tímy sa snažia poslať loptu na stranu súpera, kde ju súperi musia vrátiť. Ak lopta pritom prejde cez sieť a dopadne na zem alebo na stranu súpera, odbijajúci tím získa bod. Pokračuje sa v hre, kým niektorý z tímov nedosiahne stanovený počet bodov na víťazstvo. Hrá sa (najčastejšie) na tri víťazné sety. Maximálny počet odohraných setov je $2n-1$, kde jeden tím vyhral $n$ setov. Set sa hrá tak dlho, až jeden tím dosiahne určitý bodový zisk, spravidla je to 25 alebo 30, a musí pri tom mať najmenej o dva body viac ako druhý tím.  

Medzi najdôležitejšie pravidlá volejbalu patrí: 
1. Hrá sa (najčastejšie) na tri víťazné sety.
2. Každý tím má právo na tri dotyky lopty pred tým, ako ju musí poslať na súperovu stranu.  
3. Hráč nesmie dvakrát po sebe loptu odbiť.
4. Hráč nesmie loptu chytiť alebo niesť.  
5. Hráč môže podať z akejkoľvek časti začiatočnej čiary.
6. Hráč, ktorý sa nachádza pod sieťou, nesmie prekročiť sieť alebo dotknúť sa súperovej strany ihriska.
7. V prípade, že lopta dopadne na čiaru, považuje sa to za platný zásah.


Toto je len zopár pravidiel volejbalu, ktoré by hráči mali dodržiavať. Rozhodcovia dbajú na ich dodržiavanie. Ich dodržiavanie pomáha zabezpečiť spravodlivú a bezproblémovú hru.

Kvôli jednoduchšej interpretácii budeme označovať tím **CSUN** ako tím **A** a tím **UCLA** alebo iného súpera ako tím **B**. Dôležitými hodnotami, ktoré budeme ďalej využívať vo výpočtoch, sú odhady pravdepodobností získania bodu ak daný tím podáva. Označíme ich ako $\hat{p_A}$ a $\hat{p_B}$.  
Takéto odhady pravdepodobností spravíme pre každý zápas samostatne.

```r
# odhad pre tím A
p_A=9/43
p_A
# odhad 1-p_A, ozn. p_A2
p_A2=1-9/43
p_A2
```

```r
# odhad pre tím B
p_B=26/59
p_B
# odhad 1-p_B, ozn. p_B2
p_B2=1-26/59
p_B2
```

Ukážeme si tri pohľady, spôsoby, prístupy akými môžeme analyzovať, či už jeden volejbalový zápas alebo všetkých 55 setov. 

### 1. časť

Získané odhady pravdepodobností z tabuľky **Tab. 1** využijeme teraz. Musíme poznamenať, že pre každý zápas platí iná tabuľka. V tejto časti sa na základe vzťahov $E(R_l^A)=N_A(1-\hat{p_B})\hat{p_A}^{l-1}$ a $E(R_l^B)=N_B(1-\hat{p_A})\hat{p_B}^{l-1}$ ($N_A$ a $N_B$ ozn. počet získaných bodov zo súperovho podania) budeme snažiť zistiť, očakávaný počet šnúr danej dĺžky. Uvedené vzťahy pre stredné hodnoty pochádzajú z **geometrického rozdelenia**. Stredné hodnoty pre tím **A** a pre tím **B** spravíme pre každý zápas osobitne a nakoniec ich spočítame. Získané počty porovnáme s počtom šnúr rovnakej dĺžky, ktoré v zápasoch reálne nastali.  

$$\text{Tab. 2: Skutočný vs očakávaný počet šnúr v 55 setoch}$$

Ak si zoberieme na porovnanie druhý a tretí stĺpec vidíme, že očakávaný počet sa mierne odchyľuje od skutočného. Ale môžeme zhodnotiť, že použité vzťahy veľmi dobre odhadujú počet šnúr rôznej dĺžky. 

### 2. časť

V tejto časti použijeme odhady pravdepodobnosti získania bodu, ak daný tím servisuje, ktoré získame týmto výpočtom $\hat{p_A}=n_A/(n_A+N_B-n_B)$ a $\hat{p_B}=n_B/(n_B+N_A-n_A)$, kde $n_A$ a $n_B$ predstavuje počet získaných bodov z vlastného podania, $N_A$ a $N_B$ je celkový počet bodov daného tímu za celý set. Potom nasimulujeme 100-krát jednotlivé sety. Následne spočítame koľkokrát, nami nasimulovanom zápase, naskytli šnúry rôznych dĺžok. Potom ich vydelíme 100 pretože chceme zistiť počty na jeden zápas. A získané počty šnúr a porovnáme so skutočnými.

$$\text{Tab. 3: Skutočný vs simulovaný počet šnúr v 55 setoch}$$

Opäť môžeme vidieť, že hodnoty v druhom a v treťom stĺpci sú mierne odlišné. Mohli by sme dané hodnoty zo simulácie vylepšiť keby sme dali vyšší počet simulácii. 

### 3. časť

Hlavnou myšlienkou 3. časti je, že sme si vybrali jeden zápas, na ktorom sme zisťovali závislosť medzi počtom prerušených šnúr a bodmi, ktoré daný tím získal ich podávaním. Využili sme na to **chi-kvadrát test nezávislosti**. Bonusom v tejto časti bolo zistenie percentuálnej úspešnosti prerušenia šnúr. Tentokrát to je zápas medzi CSUN a Santa Barbara, ktorý sa konal 20.10.2007. Z tohto zápasu vieme, koľko dané tímy dokázali mať šnúr rôznych dĺžok.  

$$\text{Tab. 4: Skutočný počet šnúr v zápase CSUN vs. Santa Barbara}$$

Výpočet prebieha nasledovne:   
- zoberieme si počet šnúr dĺžky 3 pre CSUN vieme, že možnosť prerušenia bola iba 1, pre šnúru dĺžky 4 tím mal dve mžnosti a tak ďalej budeme postupovať a nakoniec to spočítame: $$4\cdot1+4\cdot2+0\cdot3+1\cdot4+1\cdot5=21$$ to je počet možností, koľko mala **Santa Barbara**, aby prerušila šnúry.     

- Úspešne sa **Santa Barbara** podarilo prerušiť $$4+4+0+1+1=10$$.  

- Takže celkova úspešnosť **Santa Barbary** v prerušovaní šnúr je $$\frac{10}{21}=48\%$$

Podobne aj pre tím **CSUN**:
- Celkový počet možností prerušenia šnúr $$6\cdot1+1\cdot2+3\cdot3+1\cdot4+0\cdot5=21$$
- Úspešný počet prerušenia šnúr $$6+1+3+1+0=11$$.
- Celkova úspešnosť **CSUN** v prerušovaní šnúr je $$\frac{11}{21}=52\%$$


```r
# možnosti prerušenia šnúr CSUN
p=4*1+4*2+0*3+1*4+1*5
p
# počet úspešne prerušených
u=4+4+0+1+1
u
# celková úspešnosť
c=u/p*100
c
```

Kde počet možností prerušenia šnúr CSUN mala Santa Barbara $p=21$. Počet úspešne prerušených šnúr Santa Barbarou $u=10$ a celková úspešnosť Santa Barbary v prerušovaní šnúr predstavuje $c=47,62 \%$.

```r
# možnosti prerušenia šnúr Santa Barbara
p=6*1+1*2+3*3+1*4+0*5
p
# počet úspešne prerušených
u=6+1+3+1+0
u
# celková úspešnosť
c=u/p*100
c
```

Počet možností prerušenia šnúr Santa Barbary mal CSUN $p=21$, počet úspešne prerušených šnúr CSUN $u=11$ a celková úspešnosť CSUN v prerušovaní šnúr vyšla $c=52,38 \%$.

Z priebehu hry vieme, že CSUN prerušila šnúry Santa Barbary s úspešnosťou $47 \%$, čo je veľmi podobné tomu, čo nám vyšlo vyššie výpočtom ($47,62 \%$). A Santa Barbara prerušila šnúry s úspešnosťou $52 \%$, opäť sa to približuje hodnote získanej vyššie ($52,38 \%$). 

  

Ďalej autor zisťoval pomocou chi-kvadrátu testu pre nezávislosť závislosť medzi úspešne prerušenými šnúrami a bodmi, ktoré tím získal zo svojho podania. Nulová hypotéza $H_0$, ktorá hovorí o tom, že úspešné prerušenia šnúr a body získané zo svojho podania sú navzájom závislé. Alternatívna hypotéze $H_1$ hovorí o tom, že skúmané situácie sú nezávislé. Keďže nemáme body, ktoré získali tímy zo svojich podaní, tak nevieme výpočet podrobne zrealizovať pri tomto zápase. Dôkladnejšie si to ukážeme pri nami analyzovanom zápase. V článku je zverejnená len $p-hodnota$. Autorovi vyšla $p-hodnota=0,81$, čo je oveľa viac ako $0,05$. Takže nulovú hypotézu nemôžeme zamietnuť.
